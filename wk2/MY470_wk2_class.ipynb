# Strings

* Ordered sequences of characters
* Immutable

```python
x = 'my string'

# Capitalises the first character (chr) of the string
x = x.capitalize()

# prints the string
print(x)

# prints the chr at index 3
print(x[3])

# prints the last chr
print(x[-1])

# print a range 
# NOTE: not inclusive of the last index, 4 chrs because python starts at 0
print(x[0:4])

# Index one to the last index
# Again, not inclusive
print(x[1:-1])

# EXTENDED SLICING 
# Get every other (2) item in the string.
print(x[::2])

# Reverse steps, every other chr
print(x[::-2])
```

```python
# Exercise 11: Print the second smallest and the second largest numbers 
# in this list of unique numbers


x = [2, 5, 0.7, 0.2, 0.1, 6, 7, 3, 1, 0, 0.3]
sorted_x = sorted(x)

sec_lar = sorted_x[-2]
sec_sma = sorted_x[1]

print(sec_lar)
print(sec_sma)
```

```python

#Another way to do this : using For Loop  

x = [2, 5, 0.7, 0.2, 0.1, 6, 7, 3, 1, 0, 0.3]
#The largest number
max_num = 0
for num_a in x:
    if num_a > max_num:
        max_num = num_a
x.remove(max_num)

#second largest number
sec_num = 0
for num_b in x:
    if num_b > sec_num:
       sec_num = num_b
print(sec_num)

    
#The smallest number
min_num = 10
for num_c in x:
    if num_c < min_num:
       min_num = num_c
x.remove(min_num)

#Second smallest number
    
sec_num = 10
for num_d in x:
    if num_d < sec_num:
       sec_num = num_d
print(sec_num)

#Second smallest number
```
